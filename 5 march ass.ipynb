Bokeh server is a Python library that allows you to create and deploy interactive Bokeh plots as web applications. Bokeh server provides a way to create dynamic, interactive plots that can be updated in real-time as data changes or as the user interacts with the plot. Bokeh server can be used to create a wide range of interactive applications, from simple dashboards to complex data exploration tools.

To use Bokeh server, you need to create a Python script that defines your interactive plot using the Bokeh library. The script then uses the bokeh.server.server module to start the Bokeh server and publish your plot as a web application. Here's an example of how to use Bokeh server to create an interactive plot:

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    plot = figure()
    plot.circle([1,2,3], [4,5,6])

    # Generate the plot HTML and JavaScript code
    plot_html, plot_js = components(plot)

    # Render the HTML template with the plot components
    return render_template('index.html', plot_html=plot_html, plot_js=plot_js)

if __name__ == '__main__':
    app.run()
```
